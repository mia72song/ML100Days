# 範例重點

如何使用亂數, 資料集來操作


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 180)
y = np.cos(x*np.pi /180.0)

plt.plot(x, y)

plt.show()
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

colors = np.random.rand(n)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.scatter(X, Y, c=colors)

plt.show()
```
